## II-Faisons le calcul

1. Construire un programme qui joue 1000 games, où une wining_door et un player_choice
sont choisis au hasard, un compteur wins est incrémenté à chaque fois que wining_door
et player_choice sont identiques, on devrait trouver autour de 333 wins

```python
from random import randint
nb_games = 1000
cmpt_wins = 0

for i in range(nb_games) :
    winning_door = randint(0,2)
    player_choice = randint(0,2)
    if winning_door == player_choice : 
        cmpt_wins += 1

print("Nombres de wins :", cmpt_wins)
# Nombres de wins 304 ! 
    
```

## II - Avec un ordinateur quantique

Le jeu est identique.

Nous allons utiliser trois qubits q0, q1 et q2 un pour chacune des portes,et sa valeur sera à |1⟩ si il s’agit de la porte gagnante

```python
from qiskit import *
import numpy as np 

nb_games = 1000
nb_choix = 3

circuit_mt = QuantumCircuit(nb_choix + 1)

circuit_mt.barrier()

# Rotation de 1.91 selon y
circuit_mt.ry(1.91,0)
circuit_mt.barrier()

#  CHadamar 
circuit_mt.ch(0,1)
circuit_mt.barrier()

# CX 1 
circuit_mt.cx(1,2)
circuit_mt.barrier()

# CX 2
circuit_mt.cx(0,1)
circuit_mt.barrier()

# X 
circuit_mt.x(0)
circuit_mt.barrier()

## choix de la porte à ouvrir

# Voiture derrière q0
circuit_mt.cx(0,3)
circuit_mt.barrier()

# Voiture derrière q1
circuit_mt.barrier()

# Voiture derrière q2
circuit_mt.ch(2,3)

# Mesure 
circuit_mt.measure_all()

circuit_mt.draw()
```

```python
from qiskit_aer import Aer
from qiskit import transpile
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Utilisation du simulateur local Aer
aer_sim = Aer.get_backend('aer_simulator')

# Transpilation du circuit pour le simulateur
optimized_circuit = transpile(circuit_mt, backend=aer_sim)

# Simulation du circuit optimisé
results = aer_sim.run(optimized_circuit).result()

# Extraction des résultats de mesure
answer = results.get_counts()

# Affichage de l'histogramme des résultats
plot_histogram(answer)
```
